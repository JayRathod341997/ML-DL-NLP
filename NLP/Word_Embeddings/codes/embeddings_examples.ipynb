# Word Embeddings - Examples

```python
# Using Gensim for Word2Vec
from gensim.models import Word2Vec
import nltk
nltk.download('punkt', quiet=True)
```

```python
# Sample corpus
sentences = [
    ['king', 'queen', 'prince', 'princess'],
    ['man', 'woman', 'boy', 'girl'],
    ['dog', 'cat', 'pet', 'animal']
]

# Train Word2Vec
model = Word2Vec(sentences, vector_size=100, window=3, min_count=1)

# Get word vector
print("King vector:", model.wv['king'][:5], "...")
print("Vector size:", len(model.wv['king']))
```

```python
# Word similarity
print("Similar to 'king':", model.wv.most_similar('king', topn=3))
print("Similar to 'dog':", model.wv.most_similar('dog', topn=3))
```

```python
# Analogy: king - man + woman
try:
    result = model.wv.most_similar(positive=['king', 'woman'], negative=['man'])
    print("king - man + woman =", result)
except:
    print("Need more data for analogy")
```

## Summary

Word embeddings capture semantic meaning in dense vectors. Similar words cluster together in vector space.
