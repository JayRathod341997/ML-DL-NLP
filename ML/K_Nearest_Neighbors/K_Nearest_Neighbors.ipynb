## 📖 Story Time: Finding Your Tribe 👫👫👫

Imagine you're at a party and you don't know anyone!

- You look at the 5 people **closest** to you
- If 4 of them are engineers, you're probably an engineer too!
- If 3 are artists, you might be more like artists

**KNN** works exactly the same way!
- Find the K nearest neighbors (K = number of neighbors to check)
- See what **category** most of them belong to
- Classify the new point with the **majority vote**!

---

## 🧠 The Simple Idea

KNN is like:
- Looking at **nearby friends** to guess who you are
- **K** = how many neighbors to check (usually an odd number)
- Voting: if 3 neighbors say "cat" and 2 say "dog", you're a cat!

```python
# Let's import what we need! 📦
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
```

---

## 🎮 Example 1: Video Game Character Classification!

```python
# Let's create game character data!
# Features: [Speed (1-10), Strength (1-10)]
# Categories: Warrior, Speedster, Wizard

X_game = np.array([
    [2, 9],   # Slow but strong → Warrior
    [3, 8],   # Slow but strong → Warrior
    [2, 7],   # Slow but strong → Warrior
    [8, 2],   # Fast but weak → Speedster
    [9, 3],   # Fast but weak → Speedster
    [7, 2],   # Fast but weak → Speedster
    [5, 5],   # Balanced → Wizard
    [4, 6],   # Balanced → Wizard
    [6, 4],   # Balanced → Wizard
])

y_game = np.array([0, 0, 0, 1, 1, 1, 2, 2, 2])  # 0=Warrior, 1=Speedster, 2=Wizard

class_names = ['⚔️ Warrior', '💨 Speedster', '🧙 Wizard']

# Train KNN with K=3
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_game, y_game)

print("🎮 Game Character Data:")
print("-" * 50)
for x, y in zip(X_game, y_game):
    print(f"   Speed={x[0]}, Strength={x[1]} → {class_names[y]}")
```

---

## 🔮 Making Predictions!

```python
# Let's classify new characters!
new_characters = np.array([
    [3, 8],   # Slow, strong
    [8, 3],   # Fast, weak  
    [5, 5],   # Balanced
])

predictions = knn_model.predict(new_characters)
proba = knn_model.predict_proba(new_characters)

print("🔮 Predictions for New Characters:")
print("-" * 50)
for char, pred, p in zip(new_characters, predictions, proba):
    print(f"\nNew Character: Speed={char[0]}, Strength={char[1]}")
    print(f"   → {class_names[pred]}")
    print(f"   → Votes: {class_names[0]}={p[0]*100:.0f}%, {class_names[1]}={p[1]*100:.0f}%, {class_names[2]}={p[2]*100:.0f}%")
```

---

## 📊 Let's Visualize KNN!

```python
# Visualize the decision boundaries!
plt.figure(figsize=(12, 8))

# Create mesh grid
x_min, x_max = X_game[:, 0].min() - 1, X_game[:, 0].max() + 1
y_min, y_max = X_game[:, 1].min() - 1, X_game[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))

Z = knn_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='RdYlBu')

# Plot training data
colors = ['blue', 'green', 'red']
for i, (name, color) in enumerate(zip(class_names, colors)):
    plt.scatter(X_game[y_game==i, 0], X_game[y_game==i, 1], 
               c=color, s=150, label=name, edgecolors='black')

# Plot new points
plt.scatter(new_characters[:, 0], new_characters[:, 1], 
           c='yellow', s=200, marker='*', label='New Characters', edgecolors='black')

plt.xlabel('Speed', fontsize=12)
plt.ylabel('Strength', fontsize=12)
plt.title('🎮 KNN Classification (K=3)', fontsize=14, fontweight='bold')
plt.legend()
plt.tight_layout()
plt.show()

print("\n⭐ Yellow stars = New characters we want to classify")
print("   Colors show which group they belong to!")
```

---

## 🔍 How KNN Makes Decisions - Step by Step!

```python
# Let's see how KNN classifies one point!
test_point = np.array([[4, 7]])

# Get the 3 nearest neighbors
distances, indices = knn_model.kneighbors(test_point)

print("🔍 How KNN classifies point (Speed=4, Strength=7):")
print("-" * 50)
print(f"\n📍 Test point: Speed=4, Strength=7")
print(f"\n👥 3 Nearest Neighbors:")
for i, (dist, idx) in enumerate(zip(distances[0], indices[0])):
    neighbor = X_game[idx]
    label = class_names[y_game[idx]]
    print(f"   {i+1}. Distance={dist:.2f}: Speed={neighbor[0]}, Strength={neighbor[1]} → {label}")

prediction = knn_model.predict(test_point)
print(f"\n🗳️ Voting Result: {class_names[prediction[0]]} wins!")
```

---

## 📈 Example 2: Comparing Different K Values!

```python
# Let's see how different K values affect the boundary!
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for ax, k in zip(axes, [1, 3, 7]):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_game, y_game)
    
    # Create mesh
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    ax.contourf(xx, yy, Z, alpha=0.3, cmap='RdYlBu')
    for i, color in enumerate(colors):
        ax.scatter(X_game[y_game==i, 0], X_game[y_game==i, 1], 
                  c=color, s=100, edgecolors='black')
    
    ax.set_title(f'K = {k}', fontsize=14, fontweight='bold')
    ax.set_xlabel('Speed')
    ax.set_ylabel('Strength')

plt.suptitle('🔢 How K Affects Decision Boundaries!', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("\n📊 Notice how:")
print("   K=1: Very wiggly boundaries (can overfit)")
print("   K=3: Balanced boundaries")
print("   K=7: Smoother boundaries (can underfit)")
```

---

## ⚠️ Important: Choosing K

| K Value | Effect |
|---------|--------|
| Small K (1-3) | Can be noisy, overfits |
| Medium K (5-10) | Usually good choice! |
| Large K | Smoother but can underfit |

### Tips for Choosing K:
- Use **odd numbers** to avoid ties
- Use **cross-validation** to find best K
- K = sqrt(n) is a common starting point

### When to Use KNN?

✅ Simple classification problems
✅ When you need interpretability
✅ When decision boundary is complex

❌ Large datasets (slow!)
❌ High-dimensional data
❌ When you need fast prediction
