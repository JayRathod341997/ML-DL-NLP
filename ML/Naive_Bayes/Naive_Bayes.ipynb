# Naive Bayes ClassifierA comprehensive guide to understanding and implementing Naive Bayes algorithm.

## What is Naive Bayes?Naive Bayes is a **probabilistic classification algorithm** based on Bayes' Theorem with the 'naive' assumption of feature independence.### Bayes' TheoremP(A|B) = P(B|A) × P(A) / P(B)In classification:P(Class|Features) = P(Features|Class) × P(Class) / P(Features)The 'naive' part assumes all features are independent, so:P(Features|Class) = P(Feature1|Class) × P(Feature2|Class) × ... × P(FeatureN|Class)

## How Does Naive Bayes Work?### Types of Naive Bayes**1. Gaussian NB** - Assumes features follow normal distribution- Used for continuous data like height, weight**2. Multinomial NB** - For discrete counts like word frequencies- Used for text/document classification**3. Bernoulli NB** - Binary features (present/absent)- Used for text classification with binary features**4. Complement NB** - Better for imbalanced datasets

## When to Use Naive Bayes- Text classification - spam detection, sentiment analysis- Multi-class classification problems- High-dimensional data- When training data is limited- Real-time prediction

## Benefits / Pros Extremely fast - both training and prediction Works well with high-dimensional data Good baseline model Handles missing values Easy to implement Provides probabilistic outputs## Cons / Limitations Naive independence assumption - rarely true Zero frequency problem - if feature not seen in training Limited expressiveness - can't capture complex relationships

## Real-World Applications- Spam Email Detection- Sentiment Analysis- Document Categorization- Medical Diagnosis- Credit Scoring- Weather Prediction

## Implementation Example

```python
from sklearn.naive_bayes import GaussianNBimport numpy as npimport matplotlib.pyplot as pltX = np.array([    [70, 22], [75, 24], [80, 25], [85, 23], [90, 26], [88, 28], [72, 23], [78, 25],    [82, 24], [76, 27],    [140, 32], [150, 35], [145, 33], [160, 38], [155, 36], [165, 40], [142, 34], [148, 37],    [158, 39], [152, 35]])y = np.array([0]*10 + [1]*10)print("Dataset:", X.shape)print("Classes: 0 = No Diabetes, 1 = Diabetes")
```

```python
model = GaussianNB()model.fit(X, y)new_patient = np.array([[100, 30]])prediction = model.predict(new_patient)probability = model.predict_proba(new_patient)[0]print(f"Prediction: {'Has Diabetes' if prediction[0] == 1 else 'No Diabetes'}")print(f"Probabilities: No Diabetes={probability[0]:.2%}, Diabetes={probability[1]:.2%}")
```

## Visualization

```python
h = 1.0x_min, x_max = X[:, 0].min() - 10, X[:, 0].max() + 10y_min, y_max = X[:, 1].min() - 3, X[:, 1].max() + 3xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))Z = model.predict(np.c_[xx.ravel(), yy.ravel()])Z = Z.reshape(xx.shape)Z_prob = model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1].reshape(xx.shape)plt.figure(figsize=(12, 8))contour = plt.contourf(xx, yy, Z_prob, levels=50, cmap='RdYlGn_r', alpha=0.8)plt.colorbar(contour, label='Probability of Diabetes')plt.contour(xx, yy, Z_prob, levels=[0.5], colors='black', linewidths=2, linestyles='--')plt.scatter(X[y==0, 0], X[y==0, 1], c='green', s=100, edgecolors='black', label='No Diabetes')plt.scatter(X[y==1, 0], X[y==1, 1], c='red', s=100, edgecolors='black', label='Diabetes')plt.scatter(new_patient[0, 0], new_patient[0, 1], c='yellow', s=300, marker='*', edgecolors='black', label='New Patient')plt.xlabel('Glucose Level', fontsize=12)plt.ylabel('BMI', fontsize=12)plt.title('Naive Bayes Classification: Diabetes Prediction', fontsize=14)plt.legend(loc='upper left')plt.grid(True, alpha=0.3)plt.tight_layout()plt.show()
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_reporty_pred = model.predict(X)print(f"Accuracy: {accuracy_score(y, y_pred):.2%}")print("\nConfusion Matrix:")print(confusion_matrix(y, y_pred))print("\nClassification Report:")print(classification_report(y, y_pred, target_names=['No Diabetes', 'Diabetes']))
```
