## 🧠 The Simple Idea

K-Means is like:
- Picking **K** center points (centroids)
- Assigning each point to the **nearest center**
- Moving centers to the **middle** of their group
- Repeating until everything is organized!

Steps:
1. Choose K (number of clusters)
2. Pick K random starting points
3. Assign each point to nearest center
4. Move centers to average position
5. Repeat until stable!

```python
# Let's import what we need! 📦
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
```

---

## 🎨 Example 1: Grouping Colors!

```python
# Let's create color data!
# Each point is [Red, Green, Blue] - like RGB values

# Let's simulate some color groups
np.random.seed(42)

# Create 3 distinct color clusters
# Cluster 1: Reds
reds = np.random.normal(loc=200, scale=30, size=(50, 3))
reds = np.clip(reds, 0, 255)

# Cluster 2: Greens
greens = np.random.normal(loc=[50, 200, 50], scale=30, size=(50, 3))
greens = np.clip(greens, 0, 255)

# Cluster 3: Blues
blues = np.random.normal(loc=[50, 50, 200], scale=30, size=(50, 3))
blues = np.clip(blues, 0, 255)

# Combine all colors
X_colors = np.vstack([reds, greens, blues])

# Shuffle the data
np.random.shuffle(X_colors)

print("🎨 Color Data Created!")
print(f"   Total colors: {len(X_colors)}")
print(f"   We want to find 3 groups (K=3)")
```

---

## 🎓 Training K-Means

```python
# Train K-Means with K=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X_colors)

# Get cluster labels and centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_

print("🎉 K-Means Clustering Complete!")
print(f"\n📊 Cluster Centers (the average color of each group):")
for i, center in enumerate(centers):
    print(f"   Cluster {i}: RGB({center[0]:.0f}, {center[1]:.0f}, {center[2]:.0f})")
```

```python
# Let's visualize the clusters!
fig = plt.figure(figsize=(16, 6))

# Plot 1: Show clusters
ax1 = fig.add_subplot(1, 2, 1)
colors_list = ['red', 'green', 'blue', 'purple', 'orange']
for i in range(3):
    cluster_points = X_colors[labels == i]
    ax1.scatter(cluster_points[:, 0], cluster_points[:, 1], 
               c=[colors_list[i]]*len(cluster_points), alpha=0, s=50)
    ax1.scatter(cluster_points[:, 0], cluster_points[:, 1], 
               c=cluster_points/255, s=50, label=f'Cluster {i}')
    
# Plot centers
ax1.scatter(centers[:, 0], centers[:, 1], c='black', s=300, 
           marker='X', edgecolors='white', linewidths=2, label='Centers')
ax1.set_xlabel('Red')
ax1.set_ylabel('Green')
ax1.set_title('🎨 Color Clusters (R vs G)', fontsize=14, fontweight='bold')
ax1.legend()

# Plot 2: Just show RGB as 3D
ax2 = fig.add_subplot(1, 2, 2)
for i in range(3):
    cluster_points = X_colors[labels == i]
    ax2.scatter(cluster_points[:, 1], cluster_points[:, 2], 
               c=cluster_points/255, s=50, label=f'Cluster {i}')
ax2.scatter(centers[:, 1], centers[:, 2], c='black', s=300, 
           marker='X', edgecolors='white', linewidths=2)
ax2.set_xlabel('Green')
ax2.set_ylabel('Blue')
ax2.set_title('🎨 Color Clusters (G vs B)', fontsize=14, fontweight='bold')
ax2.legend()

plt.tight_layout()
plt.show()

print("\n🟡 Black X marks = Cluster Centers")
```

## 🏪 Example 2: Customer Segmentation!

```python
# Let's create customer data!
# Features: [Annual Income ($K), Spending Score (1-100)]

# Create synthetic customer data with 5 segments
np.random.seed(42)

# Generate 5 different customer groups
customers = []
centers = [
    [20, 80],   # Low income, High spending
    [80, 80],   # High income, High spending
    [20, 20],   # Low income, Low spending
    [80, 20],   # High income, Low spending
    [50, 50],   # Average
]

for center in centers:
    group = np.random.normal(center, scale=10, size=(30, 2))
    customers.append(group)

X_customers = np.vstack(customers)
X_customers = np.clip(X_customers, 0, 100)

# Train K-Means with K=5
kmeans_customers = KMeans(n_clusters=5, random_state=42, n_init=10)
labels_customers = kmeans_customers.fit_predict(X_customers)
centers_customers = kmeans_customers.cluster_centers_

print("🏪 Customer Segmentation Complete!")
print(f"   Number of customers: {len(X_customers)}")
print(f"   Number of segments: 5")
```

```python
# Visualize customer segments!
plt.figure(figsize=(12, 8))

# Define colors for each segment
segment_colors = ['red', 'blue', 'green', 'purple', 'orange']
segment_names = ['Budget Spenders', 'Premium Customers', 'Careful Savers', 
                'Big Savers', 'Average Customers']

for i in range(5):
    segment = X_customers[labels_customers == i]
    plt.scatter(segment[:, 0], segment[:, 1], 
               c=segment_colors[i], s=100, label=f'Segment {i}: {segment_names[i]}',
               alpha=0.7, edgecolors='black')

# Plot centers
plt.scatter(centers_customers[:, 0], centers_customers[:, 1], 
           c='black', s=300, marker='X', edgecolors='white', 
           linewidths=2, label='Segment Centers')

plt.xlabel('💰 Annual Income ($K)', fontsize=12)
plt.ylabel('🛒 Spending Score', fontsize=12)
plt.title('🏪 Customer Segmentation with K-Means!', fontsize=14, fontweight='bold')
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n🎯 Each dot is a customer, colors show their segment!")
```

---

## 📊 How to Choose K? The Elbow Method!

```python
# Let's find the best K using the Elbow Method!

# Test different K values
k_range = range(1, 11)
inertias = []

for k in k_range:
    kmeans_test = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_test.fit(X_customers)
    inertias.append(kmeans_test.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertias, 'bo-', linewidth=2, markersize=10)
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('Inertia (Total Distance to Centers)', fontsize=12)
plt.title('📊 The Elbow Method - Finding Optimal K!', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

# Mark the elbow point
plt.axvline(x=5, color='r', linestyle='--', label='Optimal K=5')
plt.legend()

plt.tight_layout()
plt.show()

print("\n📈 Look for the 'elbow' where the line bends!")
print("   In this case, K=5 looks like a good choice!")
```

## ⚠️ Important: K-Means Limitations

### Things to Watch Out For:

| Problem | Solution |
|---------|----------|
| Must pick K beforehand | Use Elbow Method |
| Sensitive to initial centers | Use n_init (run multiple times) |
| Assumes spherical clusters | Try other algorithms for complex shapes |
| Sensitive to outliers | Preprocess data, remove outliers |

### When to Use K-Means?

✅ When you know the number of groups
✅ When clusters are roughly circular
✅ For initial data exploration

❌ When clusters have weird shapes
❌ When you don't know how many groups
