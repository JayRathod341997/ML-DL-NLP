# Principal Component Analysis (PCA)A comprehensive guide to understanding and implementing PCA for dimensionality reduction.

## What is PCA?Principal Component Analysis (PCA) is an **unsupervised learning technique** used for dimensionality reduction and data visualization.It transforms the data into a new coordinate system where:- The first principal component has the largest variance- Each subsequent component has the largest possible variance orthogonal to previous componentsThink of it as finding the 'best angle' to view your data - the angle that shows the most information.

## How Does PCA Work?### Mathematical Steps**1. Standardize the Data**- Subtract mean and divide by standard deviation- Ensures all features are on the same scale**2. Compute Covariance Matrix**- Measures how features vary together- Creates a matrix showing relationships between all feature pairs**3. Calculate Eigenvectors and Eigenvalues**- Eigenvectors: directions of maximum variance (principal components)- Eigenvalues: amount of variance in each direction**4. Select Top Components**- Sort by eigenvalues (largest first)- Choose top k components that explain desired variance**5. Transform Data**- Project data onto selected principal components

## When to Use PCA- Reduce dimensionality of high-dimensional data- Visualize data in 2D or 3D- Remove correlated features- Speed up other ML algorithms- Compress images or data- Detect outliers

## Benefits / Pros Removes correlated features effectively Reduces overfitting in ML models Speaks up computation time for other algorithms Helps visualize high-dimensional data Removes noise in data## Cons / Limitations Loses some information (by design) Components are hard to interpret Assumes linear relationships Sensitive to outliers

## Real-World Applications- Image Compression - reduce image file sizes- Facial Recognition - eigenfaces approach- Finance - portfolio optimization- Biology - gene expression analysis- Marketing - customer behavior segmentation- Quality Control - detecting defects

## Implementation Example

```python
from sklearn.decomposition import PCAfrom sklearn.preprocessing import StandardScalerimport numpy as npimport matplotlib.pyplot as pltnp.random.seed(42)n_samples = 200mean = [0, 0, 0]cov = [[1, 0.8, 0.5], [0.8, 1, 0.6], [0.5, 0.6, 1]]X = np.random.multivariate_normal(mean, cov, n_samples)print("Original data shape:", X.shape)print("Number of features:", X.shape[1])
```

```python
scaler = StandardScaler()X_scaled = scaler.fit_transform(X)pca = PCA()X_pca = pca.fit_transform(X_scaled)print("Transformed data shape:", X_pca.shape)print("\nExplained Variance Ratio per Component:")for i, var in enumerate(pca.explained_variance_ratio_):    print(f"  PC{i+1}: {var:.4f} ({var*100:.2f}%)")total_var = sum(pca.explained_variance_ratio_)print(f"\nTotal variance explained: {total_var:.4f}")
```

## Visualization

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))axes[0].scatter(X[:, 0], X[:, 1], alpha=0.6)axes[0].set_xlabel('Feature 1')axes[0].set_ylabel('Feature 2')axes[0].set_title('Original Data (2 Features)')axes[0].grid(True, alpha=0.3)axes[1].scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.6)axes[1].set_xlabel('First Principal Component')axes[1].set_ylabel('Second Principal Component')axes[1].set_title('Transformed Data (2 Components)')axes[1].grid(True, alpha=0.3)plt.tight_layout()plt.show()
```

```python
plt.figure(figsize=(10, 5))components = range(1, len(pca.explained_variance_ratio_) + 1)plt.bar(components, pca.explained_variance_ratio_, alpha=0.7, label='Individual')plt.plot(components, np.cumsum(pca.explained_variance_ratio_), 'ro-', label='Cumulative')plt.xlabel('Principal Component')plt.ylabel('Explained Variance Ratio')plt.title('Variance Explained by Each Component')plt.legend()plt.grid(True, alpha=0.3)plt.tight_layout()plt.show()
```
