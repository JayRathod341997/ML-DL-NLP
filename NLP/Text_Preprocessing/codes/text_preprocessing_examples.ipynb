## 1. Tokenization

```python
from nltk.tokenize import word_tokenize, sent_tokenize

text = "Natural Language Processing is fascinating! It helps computers understand human language."

# Word Tokenization
word_tokens = word_tokenize(text)
print("Word Tokens:", word_tokens)

# Sentence Tokenization
sent_tokens = sent_tokenize(text)
print("\nSentence Tokens:", sent_tokens)
```

## 2. Stemming

```python
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer

words = ["running", "runs", "runner", "ran", "easily", "studies", "studying"]

porter = PorterStemmer()
snowball = SnowballStemmer("english")
lancaster = LancasterStemmer()

print("Word      | Porter    | Snowball  | Lancaster")
print("-" * 50)
for word in words:
    print(f"{word:10} | {porter.stem(word):10} | {snowball.stem(word):10} | {lancaster.stem(word)}")
```

## 3. Lemmatization

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["running", "ran", "better", "good", "children", "ate", "eating"]

print("Word      | Lemma (v) | Lemma (n)")
print("-" * 40)
for word in words:
    verb_lemma = lemmatizer.lemmatize(word, 'v')  # as verb
    noun_lemma = lemmatizer.lemmatize(word, 'n')  # as noun
    print(f"{word:10} | {verb_lemma:10} | {noun_lemma}")
```

## 4. Stopword Removal

```python
from nltk.corpus import stopwords

text = "The quick brown fox jumps over the lazy dog"
tokens = word_tokenize(text.lower())

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

print("Original:", tokens)
print("After Stopword Removal:", filtered_tokens)
```

## 5. Text Normalization

```python
import re

def normalize_text(text):
    # Lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # Remove HTML tags
    text = re.sub(r'<[^>]+>', '', text)
    
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

text = "Check out https://example.com for more info! <b>Hello</b> World 123!!!"
print("Original:", text)
print("Normalized:", normalize_text(text))
```

## 6. Handling Contractions

```python
contractions_map = {
    "n't": " not",
    "'re": " are",
    "'s": " is",
    "'d": " would",
    "'ll": " will",
    "'ve": " have",
    "'m": " am"
}

def expand_contractions(text):
    for contraction, expansion in contractions_map.items():
        text = text.replace(contraction, expansion)
    return text

text = "I don't know what's happening. It's great! You'll love it."
print("Original:", text)
print("Expanded:", expand_contractions(text))
```

## 7. POS Tagging

```python
from nltk import pos_tag

text = "The quick brown fox jumps over the lazy dog"
tokens = word_tokenize(text)
pos_tags = pos_tag(tokens)

print("POS Tags:")
for word, tag in pos_tags:
    print(f"  {word:10} -> {tag}")
```

## 8. Complete Preprocessing Pipeline

```python
def preprocess_text(text, remove_stopwords=True):
    # Expand contractions
    text = expand_contractions(text)
    
    # Lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # Remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove stopwords
    if remove_stopwords:
        stop_words = set(stopwords.words('english'))
        # Keep negation words for sentiment
        stop_words.discard('not')
        stop_words.discard('no')
        tokens = [t for t in tokens if t not in stop_words]
    
    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    
    return tokens

text = "I didn't like the movie! It's terrible. http://example.com"
print("Original:", text)
print("Processed:", preprocess_text(text))
print("\nWith Stopwords:", preprocess_text(text, remove_stopwords=False))
```

## Using spaCy (Production-Grade)

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Apple is looking at buying U.K. startup for $1 billion"
doc = nlp(text)

# Tokenization
print("Tokens:", [token.text for token in doc])

# Lemmatization
print("Lemmas:", [token.lemma_ for token in doc])

# POS Tagging
print("POS Tags:", [(token.text, token.pos_, token.tag_) for token in doc])

# Named Entities
print("Entities:", [(ent.text, ent.label_) for ent in doc.ents])
```

## Using TextBlob (Simple API)

```python
from textblob import TextBlob

text = "Natural Language Processing is fascinating!"
blob = TextBlob(text)

print("Words:", blob.words)
print("Sentences:", blob.sentences)
print("Tags:", blob.tags)
print("Noun Phrases:", blob.noun_phrases)
print("Lemmas:", [word.lemmatize() for word in blob.words])
```
