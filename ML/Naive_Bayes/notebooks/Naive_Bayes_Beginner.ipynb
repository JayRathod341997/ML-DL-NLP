# Naive Bayes - Beginner Guide

## What is Naive Bayes?
Naive Bayes is a probabilistic classifier based on Bayes' theorem. It assumes all features are independent.

**Use Case:** Text classification, spam detection

```python
# Import libraries
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import numpy as np

# Sample data
X = np.array([[1, 1], [2, 2], [3, 1], [4, 2], [5, 3]])
y = np.array([0, 0, 1, 1, 1])

# Create and train model
model = GaussianNB()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Evaluate
print(f"Accuracy: {accuracy_score(y, y_pred):.2f}")

# Predict probability
prob = model.predict_proba(X)
print(f"\nClass Probabilities (Class 0, Class 1):\n{prob}")
```
