# K-Means Clustering - Beginner Guide

## What is K-Means Clustering?
K-Means groups similar data points into K clusters by minimizing distance within clusters.

**Use Case:** Customer segmentation, data grouping (unsupervised)

```python
# Import libraries
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array([[1, 1], [2, 2], [8, 8], [9, 9], [10, 10]])

# Create and train model
model = KMeans(n_clusters=2, random_state=42)
labels = model.fit_predict(X)

# Visualize
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], 
            c='red', marker='X', s=200, label='Centroids')
plt.legend()
plt.title('K-Means Clustering')
plt.show()

print(f"Cluster assignments: {labels}")
print(f"Inertia: {model.inertia_:.2f}")
```
