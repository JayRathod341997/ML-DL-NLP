# Random Forest - Beginner Guide

## What is Random Forest?
Random Forest is an ensemble learning algorithm that combines multiple decision trees to make more accurate predictions.

**Why:** Better accuracy and prevents overfitting compared to single decision tree

```python
# Import libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# Sample data
X = np.array([[1, 1], [2, 2], [3, 1], [4, 2], [5, 3]])
y = np.array([0, 0, 1, 1, 1])

# Create and train model
model = RandomForestClassifier(n_estimators=10, random_state=42)
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Evaluate
print(f"Accuracy: {accuracy_score(y, y_pred):.2f}")
print(f"Feature Importances: {model.feature_importances_}")
```
