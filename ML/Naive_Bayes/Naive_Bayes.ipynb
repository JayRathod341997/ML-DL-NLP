## 📖 Story Time: The Spam Filter! 📧

Imagine you're checking emails and trying to spot spam!

You might think:
- "FREE" in the subject → probably spam
- "Mom" in the subject → probably real
- "Buy now!" → probably spam

**Naive Bayes** does exactly this mathematically!
- It calculates the **probability** of being spam
- Based on the **words** in the email
- It's fast and surprisingly accurate!

---

## 🧠 The Simple Idea

Naive Bayes is like:
- Counting how often each word appears in spam vs. real emails
- Using those counts to **calculate probabilities**
- Combining all word probabilities to make a final decision

The "Naive" part:
- It assumes **each word is independent** (naive assumption!)
- In reality, words relate to each other
- But this assumption makes calculations MUCH faster
- And it still works great!

```python
# Let's import what we need! 📦
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB, MultinomialNB, CategoricalNB
from sklearn.feature_extraction.text import CountVectorizer
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
```

---

## 📧 Example 1: Spam Email Detection!

```python
# Let's create some email data!

# Sample emails (text)
emails = [
    "free money win now",           # Spam
    "buy cheap pills online",       # Spam
    "congratulations you won prize", # Spam
    "meeting tomorrow at office",   # Not Spam
    "project deadline approaching", # Not Spam
    "please review the document",   # Not Spam
    "free offer limited time",      # Spam
    "schedule meeting friday",      # Not Spam
]

# Labels: 1 = Spam, 0 = Not Spam
labels = np.array([1, 1, 1, 0, 0, 0, 1, 0])

# Convert text to numbers using CountVectorizer
vectorizer = CountVectorizer()
X_emails = vectorizer.fit_transform(emails).toarray()

print("📧 Email Data (converted to numbers):")
print("-" * 60)
feature_names = vectorizer.get_feature_names_out()
print(f"Features (words): {list(feature_names)}")
print()
for email, label in zip(emails, labels):
    status = "🚨 SPAM" if label == 1 else "✅ Not Spam"
    print(f"   '{email}' → {status}")
```

---

## 🎓 Training the Naive Bayes Model

```python
# Train Multinomial Naive Bayes (good for text/word counts)
nb_model = MultinomialNB()
nb_model.fit(X_emails, labels)

print("🎉 Naive Bayes Model Trained!")
print(f"\n📊 Class probabilities:")
print(f"   P(Not Spam) = {nb_model.class_prior_[0]*100:.1f}%")
print(f"   P(Spam) = {nb_model.class_prior_[1]*100:.1f}%")
```

---

## 🔮 Making Predictions!

```python
# Let's test on new emails!

new_emails = [
    "free money now",
    "meeting on monday",
    "buy cheap pills",
    "review document please",
]

# Convert new emails to numbers
X_new = vectorizer.transform(new_emails).toarray()

# Predict
predictions = nb_model.predict(X_new)
proba = nb_model.predict_proba(X_new)

print("🔮 Predictions for New Emails:")
print("-" * 60)
for email, pred, p in zip(new_emails, predictions, proba):
    status = "🚨 SPAM" if pred == 1 else "✅ Not Spam"
    print(f"\n📧 '{email}'")
    print(f"   → {status}")
    print(f"   → Confidence: Not Spam={p[0]*100:.1f}%, Spam={p[1]*100:.1f}%")
```

---

## 📊 Understanding the Math (Optional!)

Let's see how Naive Bayes calculates probabilities!

```python
# Let's trace through one example

test_email = "free money"
X_test = vectorizer.transform([test_email]).toarray()

# Get probabilities for each feature
feature_log_prob = nb_model.feature_log_prob_

print("📊 How Naive Bayes sees 'free money':")
print("-" * 60)
print(f"\nFor word 'free':")
print(f"   P(free|Not Spam) = {np.exp(feature_log_prob[0][0]):.4f}")
print(f"   P(free|Spam) = {np.exp(feature_log_prob[1][0]):.4f}")

print(f"\nFor word 'money':")
print(f"   P(money|Not Spam) = {np.exp(feature_log_prob[0][2]):.4f}")
print(f"   P(money|Spam) = {np.exp(feature_log_prob[1][2]):.4f}")

# Final prediction
final_pred = nb_model.predict(X_test)[0]
final_prob = nb_model.predict_proba(X_test)[0]

print(f"\n🗳️ Final Vote:")
print(f"   P(Not Spam|free,money) = {final_prob[0]*100:.2f}%")
print(f"   P(Spam|free,money) = {final_prob[1]*100:.2f}%")
print(f"\n   → Prediction: {'🚨 SPAM' if final_pred == 1 else '✅ Not Spam'}")
```

## 🏥 Example 2: Medical Diagnosis!

```python
# Let's use Gaussian Naive Bayes for numerical data!

# Medical data: [Fever (0-10), Cough (0-10), Fatigue (0-10)]
# Diagnosis: Flu or Cold

X_medical = np.array([
    [8, 7, 8],   # High fever, cough, fatigue → Flu
    [9, 8, 9],   # High symptoms → Flu
    [7, 6, 7],   # Flu
    [8, 7, 8],   # Flu
    [2, 5, 3],   # Low fever, some cough → Cold
    [3, 6, 2],   # Cold
    [2, 4, 3],   # Cold
    [3, 5, 2],   # Cold
])

y_medical = np.array([1, 1, 1, 1, 0, 0, 0, 0])  # 1=Flu, 0=Cold

# Train Gaussian Naive Bayes
gnb = GaussianNB()
gnb.fit(X_medical, y_medical)

# Test on new patient
new_patient = np.array([[6, 5, 6]])
prediction = gnb.predict(new_patient)
prob = gnb.predict_proba(new_patient)

print("🏥 Medical Diagnosis with Naive Bayes:")
print("-" * 50)
print(f"\nNew Patient: Fever={new_patient[0][0]}, Cough={new_patient[0][1]}, Fatigue={new_patient[0][2]}")
print(f"\n   → Diagnosis: {'😷 FLU' if prediction[0] == 1 else '🤧 COLD'}")
print(f"   → Confidence: Cold={prob[0][0]*100:.1f}%, Flu={prob[0][1]*100:.1f}%")
```

## ⚠️ Important: Types of Naive Bayes

| Type | Best For | Example |
|-----|---------|--------|
| **Multinomial** | Word counts/text | Spam detection |
| **Gaussian** | Continuous numbers | Medical data |
| **Categorical** | Categories | Survey responses |

### When to Use Naive Bayes?

✅ Very fast (great for large datasets)
✅ Works well with text
✅ Handles many features well
✅ Good baseline model

❌ Assumes feature independence (can be wrong)
❌ Not as accurate as complex models
