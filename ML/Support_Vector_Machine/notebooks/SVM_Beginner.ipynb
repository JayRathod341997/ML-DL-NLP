# Support Vector Machine (SVM) - Beginner Guide

## What is SVM?
SVM finds the best boundary (hyperplane) to separate different classes of data.

**Use Case:** Classification for both linear and non-linear problems

```python
# Import libraries
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import numpy as np

# Sample data
X = np.array([[1, 1], [2, 2], [3, 1], [4, 2], [5, 3]])
y = np.array([0, 0, 1, 1, 1])

# Create and train model
model = SVC(kernel='rbf')
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Evaluate
print(f"Accuracy: {accuracy_score(y, y_pred):.2f}")
print(f"Support Vectors: {len(model.support_vectors_)}")
```
