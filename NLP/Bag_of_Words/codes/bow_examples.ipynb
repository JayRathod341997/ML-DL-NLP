# Bag of Words (BoW) - In-Memory Examples

This notebook demonstrates Bag of Words implementation using Python.

```python
# Import libraries
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.naive_bayes import MultinomialNB
import pandas as pd
```

## 1. Basic Bag of Words with CountVectorizer

```python
# Simple corpus
corpus = [
    "I love dogs",
    "I love cats",
    "I love both cats and dogs",
    "Dogs are great pets"
]

# Create CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform
X = vectorizer.fit_transform(corpus)

# Get vocabulary
vocabulary = vectorizer.get_feature_names_out()

print("Vocabulary:", vocabulary)
print("\nBoW Matrix:")
print(X.toarray())
```

## 2. Understanding the Output

```python
# Show as DataFrame
df_bow = pd.DataFrame(X.toarray(), columns=vocabulary)
print("Documents as BoW vectors:")
for i, doc in enumerate(corpus):
    print(f"Doc {i+1}: {doc}")
    print(df_bow.iloc[i].to_dict())
    print()
```

## 3. Binary BoW

```python
# Binary BoW
binary_vectorizer = CountVectorizer(binary=True)
X_binary = binary_vectorizer.fit_transform(corpus)

print("Binary BoW Matrix:")
print(X_binary.toarray())
```

## 4. Using N-grams

```python
# With n-grams (unigrams + bigrams)
ngram_vectorizer = CountVectorizer(ngram_range=(1, 2))
X_ngram = ngram_vectorizer.fit_transform(corpus)

print("Vocabulary with bigrams:")
print(ngram_vectorizer.get_feature_names_out())
```

## 5. Vocabulary Control

```python
# Limit vocabulary
limited_vectorizer = CountVectorizer(
    max_features=5,
    min_df=2,
    max_df=0.8
)

X_limited = limited_vectorizer.fit_transform(corpus)

print("Limited Vocabulary:", limited_vectorizer.get_feature_names_out())
```

## 6. Document Similarity

```python
# New corpus
corpus_sim = [
    "I love Python programming",
    "Python is a great language",
    "I hate bugs in code",
    "Java is also popular"
]

vectorizer_sim = CountVectorizer()
X_sim = vectorizer_sim.fit_transform(corpus_sim)

# Cosine similarity
similarity = cosine_similarity(X_sim)
print("Similarity matrix (Doc 0 vs others):")
print(similarity[0])
```

## 7. Spam Detection Example

```python
# Simple spam dataset
emails = [
    "Get free money now",
    "Congratulations you won prize",
    "Meeting at 3pm tomorrow",
    "Please review document",
    "Click here to claim reward",
    "Project deadline Friday",
    "Limited time offer buy now",
    "Can we schedule call"
]

labels = [1, 1, 0, 0, 1, 0, 1, 0]

vectorizer_spam = CountVectorizer()
X_spam = vectorizer_spam.fit_transform(emails)

clf = MultinomialNB()
clf.fit(X_spam, labels)

# Test
new_emails = ["Free iPhone click now", "Team meeting tomorrow"]
X_new = vectorizer_spam.transform(new_emails)
predictions = clf.predict(X_new)

print("Spam Detection:")
for email, pred in zip(new_emails, predictions):
    print(f"  {email} -> {'SPAM' if pred==1 else 'NOT SPAM'}")
```

## Summary

Bag of Words is a simple text vectorization technique that converts documents into numerical vectors based on word counts.
