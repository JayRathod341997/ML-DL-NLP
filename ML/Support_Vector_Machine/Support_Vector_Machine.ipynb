# Support Vector Machine (SVM)A comprehensive guide to understanding and implementing SVM algorithm.

## What is SVM?Support Vector Machine (SVM) is a **powerful supervised learning algorithm** used for both classification and regression. It finds the optimal hyperplane that best separates different classes.### Key Concepts:**Hyperplane**: A decision boundary that separates classes- In 2D: a line- In 3D: a plane- In higher dimensions: a hyperplane**Support Vectors**: The data points closest to the hyperplane- These points define the margin- They are the most critical points**Margin**: The distance between the hyperplane and support vectors- SVM maximizes this margin- Larger margin = better generalization

## How Does SVM Work?### Types of SVM Kernels**1. Linear Kernel** - For linearly separable data- Simple, fast- Works when classes can be separated by a straight line**2. Polynomial Kernel** - For curved boundaries- Creates polynomial decision boundaries- More flexible than linear**3. RBF (Radial Basis Function) Kernel** - Default choice- Most popular kernel- Creates complex, non-linear boundaries- One parameter: gamma**4. Sigmoid Kernel** - Similar to neural networks- Less commonly used### The Optimization ProblemMaximize: 2/||w||Subject to: yi(w·xi + b) >= 1 for all iThis ensures correct classification with maximum margin.

## When to Use SVM- Binary and multi-class classification- High-dimensional data- When decision boundary is complex- Text classification- Image recognition- Bioinformatics

## Benefits / Pros Effective in high dimensions Works well with clear margin of separation Memory efficient (uses support vectors only) Versatile with different kernels Good generalization ability## Cons / Limitations Doesn't scale well to large datasets Doesn't perform well with noisy data Requires feature scaling Difficult to choose optimal kernel Doesn't provide probability estimates directly

## Real-World Applications- Face Detection - identifying faces in images- Text Categorization - spam detection, sentiment analysis- Handwriting Recognition - OCR applications- Medical Diagnosis - disease classification- Bioinformatics - protein classification- Stock Market - trend prediction
