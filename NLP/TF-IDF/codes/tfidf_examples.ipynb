# TF-IDF - In-Memory Examples

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

## 1. Basic TF-IDF

```python
corpus = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "cats and dogs are great"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("\nTF-IDF Matrix:")
print(np.round(X.toarray(), 3))
```

## 2. With N-grams

```python
vectorizer_ngram = TfidfVectorizer(ngram_range=(1, 2))
X_ngram = vectorizer_ngram.fit_transform(corpus)

print("Vocabulary with bigrams:", vectorizer_ngram.get_feature_names_out())
```

## 3. Document Similarity

```python
docs = [
    "Python programming is great",
    "I love coding in Python",
    "Java is also a programming language"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)

similarity = cosine_similarity(X)
print("Similarity matrix:")
print(np.round(similarity, 3))
```

## 4. Keyword Extraction

```python
doc = "Machine learning is a subset of artificial intelligence"
vectorizer = TfidfVectorizer()
vectorizer.fit([doc])

feature_names = vectorizer.get_feature_names_out()
tfidf_scores = vectorizer.transform([doc]).toarray()[0]

keywords = sorted(zip(feature_names, tfidf_scores), key=lambda x: x[1], reverse=True)
print("Keywords by TF-IDF:", keywords)
```

## 5. Sublinear TF

```python
corpus = ["word word word python python", "word python"]

tf_normal = TfidfVectorizer(sublinear_tf=False)
tf_sublinear = TfidfVectorizer(sublinear_tf=True)

X_normal = tf_normal.fit_transform(corpus)
X_sublinear = tf_sublinear.fit_transform(corpus)

print("Normal TF:", np.round(X_normal.toarray(), 3))
print("Sublinear TF:", np.round(X_sublinear.toarray(), 3))
```

## Summary

TF-IDF weights words by importance using Term Frequency and Inverse Document Frequency.
