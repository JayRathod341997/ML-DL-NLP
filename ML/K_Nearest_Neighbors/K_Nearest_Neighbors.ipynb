# K-Nearest Neighbors (KNN)

A comprehensive guide to understanding and implementing K-Nearest Neighbors algorithm.

## What is K-Nearest Neighbors?

K-Nearest Neighbors (KNN) is a **simple, instance-based, supervised learning algorithm** that can be used for both classification and regression problems.

The core idea is deceptively simple:
- **'You are who your neighbors are'**
- To classify a new point, look at its K nearest neighbors
- The new point takes the majority class (for classification) or average value (for regression)

It's called 'lazy learning' because it doesn't learn a model - it just memorizes the training data.

## How Does KNN Work?

### The Algorithm Steps:

**Step 1: Choose K**
- Select the number of neighbors (K)
- K is usually an odd number to avoid ties
- Common choices: 3, 5, 7, 11

**Step 2: Calculate Distance**
- For a new point, calculate distance to all training points
- Most common: Euclidean distance
```
d(x, y) = √((x₁-y₁)² + (x₂-y₂)² + ... + (xₙ-yₙ)²)
```

**Step 3: Find K Nearest Neighbors**
- Sort all points by distance
- Select the K closest points

**Step 4: Make Prediction**
- **Classification**: Majority vote (most common class among neighbors)
- **Regression**: Average of the values of K neighbors

### Distance Metrics:
- **Euclidean**: Straight line distance (most common)
- **Manhattan**: Sum of absolute differences
- **Minkowski**: Generalized form
- **Cosine**: Angle between vectors

## When to Use KNN

KNN is ideal for:

- **Simple classification problems** where decision boundaries are complex
- **Multi-class classification** problems
- **Recommender systems** - 'users who bought this also bought...'
- **Pattern recognition** in images or text
- **Data with local structure** - when similar items cluster together

## Benefits / Pros

 **Simple to understand and implement** - No complex mathematics

 **No training phase** - Just stores the data (lazy learning)

 **Naturally handles multi-class classification** - No special modifications needed

 **No assumptions about data** - Works without assuming any distribution

 **Can be used for both classification and regression**


## Cons / Limitations

 **Slow for large datasets** - Must compute distance to all training points

 **Curse of dimensionality** - Performance degrades with many features

 **Sensitive to irrelevant features** - All features treated equally

 **Sensitive to scale** - Features must be normalized/scaled

 **Imbalanced data issues** - May predict majority class more often

 **Missing value sensitivity** - Cannot handle missing values directly

## Real-World Applications

- **Recommendation Systems**: Netflix, Amazon product suggestions
- **Image Recognition**: Classifying handwritten digits
- **Medical Diagnosis**: Predicting diseases based on symptoms
- **Credit Scoring**: Determining loan eligibility
- **Stock Market Prediction**: Price forecasting based on historical patterns
- **Agriculture**: Crop disease identification
- **Video Recommendation**: YouTube, Netflix viewing patterns

## Implementation Example

````python
# Import necessary libraries
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
# Example: Classifying Iris flowers based on petal length and width
# Using a simplified version of the famous Iris dataset

# Class 0: Setosa-like (small petals)
X_class0 = np.array([
    [1.4, 0.2], [1.5, 0.3], [1.3, 0.2], [1.6, 0.25], [1.4, 0.2],
    [1.5, 0.2], [1.3, 0.3], [1.55, 0.25], [1.4, 0.15], [1.45, 0.25]
])

# Class 1: Versicolor-like (medium petals)
X_class1 = np.array([
    [4.5, 1.5], [4.7, 1.6], [4.3, 1.3], [4.6, 1.4], [4.8, 1.5],
    [4.4, 1.2], [4.9, 1.5], [4.5, 1.3], [4.7, 1.4], [4.6, 1.5]
])

# Class 2: Virginica-like (large petals)
X_class2 = np.array([
    [6.0, 2.2], [6.1, 2.4], [5.8, 2.0], [6.3, 2.5], [6.0, 2.1],
    [6.2, 2.3], [5.9, 2.1], [6.4, 2.4], [6.1, 2.2], [6.0, 2.0]
])

# Combine data
X = np.vstack([X_class0, X_class1, X_class2])
y = np.array([0]*10 + [1]*10 + [2]*10)

print("Dataset:")
print("Shape:", X.shape)
print("Classes:", np.unique(y, return_counts=True))
````

````python
# Scale features (important for KNN!)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_scaled, y)

# Predict for a new flower
new_flower = np.array([[5.5, 1.8]])  # Petal length, width
new_flower_scaled = scaler.transform(new_flower)
prediction = knn.predict(new_flower_scaled)

class_names = ['Setosa', 'Versicolor', 'Virginica']
print(f"New flower (Petal: {new_flower[0][0]}cm, {new_flower[0][1]}cm)")
print(f"Predicted class: {class_names[prediction[0]]}")

# Get probabilities
proba = knn.predict_proba(new_flower_scaled)[0]
print(f"\nProbabilities: Setosa={proba[0]:.2%}, Versicolor={proba[1]:.2%}, Virginica={proba[2]:.2%}")
````

## Visualization

````python
# Visualize decision boundaries
import numpy as np
from matplotlib.colors import ListedColormap

# Create mesh grid
h = 0.02  # step size
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict for each point in mesh
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.figure(figsize=(12, 8))
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

plt.contourf(xx, yy, Z, alpha=0.4, cmap=cmap_light)
plt.contour(xx, yy, Z, colors='k', linewidths=0.5, alpha=0.5)

scatter = plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap=cmap_bold, 
                      s=100, edgecolors='black', linewidth=1)

# Mark the new flower
plt.scatter(new_flower_scaled[0, 0], new_flower_scaled[0, 1], 
           c='yellow', s=300, marker='*', edgecolors='black', linewidths=2,
           label='New Flower')

plt.xlabel('Petal Length (standardized)', fontsize=12)
plt.ylabel('Petal Width (standardized)', fontsize=12)
plt.title('KNN Decision Boundaries (K=5)', fontsize=14)
plt.colorbar(scatter, label='Class')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

## Model Evaluation

````python
# Evaluate the model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = knn.predict(X_scaled)

print(f"Accuracy: {accuracy_score(y, y_pred):.2%}")

print("\nConfusion Matrix:")
print(confusion_matrix(y, y_pred))

print("\nClassification Report:")
print(classification_report(y, y_pred, target_names=class_names))

# Find optimal K
print("\n--- Finding Optimal K ---")
accuracies = []
k_range = range(1, 16)
for k in k_range:
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_scaled, y)
    accuracies.append(accuracy_score(y, knn_temp.predict(X_scaled)))

optimal_k = list(k_range)[np.argmax(accuracies)]
print(f"Best K: {optimal_k} with accuracy {max(accuracies):.2%}")
````
