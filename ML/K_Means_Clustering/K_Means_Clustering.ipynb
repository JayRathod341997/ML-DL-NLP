# K-Means ClusteringA comprehensive guide to understanding and implementing K-Means clustering algorithm.

## What is K-Means Clustering?K-Means is an **unsupervised learning algorithm** used for clustering data into K distinct groups. Unlike supervised learning, we don't have labeled output here - the algorithm discovers patterns on its own.The goal is to:1. Partition n observations into k clusters2. Minimize within-cluster variance (sum of squared distances)3. Each point belongs to the cluster with the nearest centroid

## How Does K-Means Work?### The Algorithm Steps:**Step 1: Initialize**- Choose K random points as initial centroids (cluster centers)- ( sklearn uses 'k-means++' initialization by default for better convergence)**Step 2: Assign Points to Clusters**- For each data point, calculate distance to all K centroids- Assign the point to the nearest centroid's cluster**Step 3: Update Centroids**- Calculate the mean (center) of all points in each cluster- Move the centroid to this new center**Step 4: Repeat**- Repeat steps 2-3 until convergence (centroids don't move significantly)- Or until maximum iterations are reached### Key Formulas:```Distance (Euclidean): d(x, y) = √(Σ(xi - yi)²)Inertia (Within-cluster sum of squares): Σ||x - μ||²```

## When to Use K-MeansK-Means is ideal for:- **Market Segmentation**: Grouping customers based on purchasing behavior- **Document Clustering**: Organizing documents by topic- **Image Compression**: Reducing colors in images- **Anomaly Detection**: Finding unusual patterns- **Clustering sensor data**: Grouping similar readings

## Benefits / Pros **Simple and easy to understand** - Easy to implement and explain **Scalable** - Works well with large datasets **Fast** - Time complexity is O(n * k * t) where t is iterations **Guarantees convergence** - Will always find a solution **Adapts to new data easily** - Can assign new points to existing clusters## Cons / Limitations **Must choose K manually** - Need to know the number of clusters beforehand **Sensitive to initialization** - Different initial centroids can give different results **Assumes spherical clusters** - Doesn't work well with elongated or irregular shapes **Affected by outliers** - Extreme values can distort centroids **Converges to local optima** - May not find the best clustering

## Real-World Applications- **Customer Segmentation**: Grouping customers by age, income, spending habits- **Image Segmentation**: Grouping pixels in an image for object detection- **Document Clustering**: Organizing news articles or research papers by topic- **Fraud Detection**: Identifying unusual transaction patterns- **Inventory Management**: Categorizing products for retail analysis- **Climate Patterns**: Grouping similar weather observations
