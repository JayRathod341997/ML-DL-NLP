# K-Nearest Neighbors (KNN) - Beginner Guide

## What is KNN?
KNN predicts a class by finding the K nearest data points and using their majority class.

**Key Parameter:** k (number of neighbors to consider, usually 3-7)

```python
# Import libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# Sample data
X = np.array([[1, 1], [2, 2], [3, 1], [4, 2], [5, 3]])
y = np.array([0, 0, 1, 1, 1])

# Create and train model
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Evaluate
print(f"Accuracy: {accuracy_score(y, y_pred):.2f}")

# Predict for new data
new_point = np.array([[4.5, 2.5]])
prediction = model.predict(new_point)
print(f"Prediction for {new_point[0]}: {prediction[0]}")
```
